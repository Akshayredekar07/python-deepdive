```python
from rich.progress import track
import time

epochs = 5
for epoch in track(range(epochs), description="Training epochs"):
    for batch in track(range(200), description=f"Epoch {epoch+1} batches"):
        time.sleep(0.005)
```

```python
from rich.console import Console
from rich.progress import track
import time

console = Console()

epochs = 5
for epoch in track(range(epochs), description="Training epochs", console=console):
    for batch in track(range(200), description=f"Epoch {epoch+1} batches", console=console):
        time.sleep(0.005)
```

```python
import click

def process_items(count: int, console: Console):
    for i in track(range(count), description="Processing"):
        time.sleep(0.2)
    console.print("[green]Done[/green]")

# CLI wrapper
@click.command()
@click.option("--count", default=5)
def process(count):
    process_items(count, Console())
```
