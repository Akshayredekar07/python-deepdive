```python
def reverse_brute(s: str) -> str:
    result = ""
    for i in range(len(s)-1, -1, -1):
        result += s[i]
    return result
```

```python
print(reverse_brute("bosidian"))
```

```python
def reverse_better(s: str) -> str:
    chars = list(s)
    left, right = 0, len(chars)-1
    while left < right:
        chars[left], chars[right] = chars[right], chars[left]
        left += 1
        right -= 1
    return "".join(chars)


reverse_better("bosidianvault")
```

```python
def reverse_string(s: str) -> str:
    return s[::-1]

print(reverse_string("prettier"))
```

```python
def is_palin(s: str) -> bool:
    return s == s[::-1]

is_palin("racecar")
```

```python
def is_palindrome(s: str) -> bool:
    cleaned = "".join(ch.lower() for ch in s if s.isalnum())
    left, right = 0, len(cleaned)-1
    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1

    return True
```

```python
is_palindrome("A man, a plan, a canal: Panama")
```

```python
is_palindrome("race a car")
```

```python
def check_anagrams(s1: str, s2: str) -> bool:
    return sorted(s1) == sorted(s2)


check_anagrams("listen", "silent")
```

```python
def are_anagrams_dict(s1: str, s2: str) -> bool:
    if len(s1) != len(s2):
        return False

    freq: dict[str, int] = {}
    for ch in s1:
        freq[ch] = freq.get(ch, 0) + 1
    for ch in s2:
        freq[ch] = freq.get(ch, 0) - 1

        if freq[ch] < 0:
            return False

    return True
```

```python
are_anagrams_dict("listen", "silent")
```

```python
from collections import Counter

def are_anagrams(s1: str, s2: str)-> bool:
    return Counter(s1) == Counter(s2)


are_anagrams("listen", "silent")
```

```python
are_anagrams("hello",  "world")
```
