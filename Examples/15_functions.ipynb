```python
def withdraw(balance, amount):
    if amount <= 0:
        return "amount must be positive"
    if amount > balance:
        return "insufficient funds"
    balance -= amount
    return balance

print(withdraw(100, 20))   # 80
print(withdraw(100, 200))  # insufficient funds
```

```python
def fib(n, cache={}):
    if n in cache:
        return cache[n]
    if n < 2:
        result = n
    else:
        result = fib(n - 1) + fib(n - 2)
    cache[n] = result
    return result

print(fib(10))   # 55
```

```python
def fib(n, cache=None):
    if cache is None:
        cache = {}
    if n in cache:
        return cache[n]
    if n < 2:
        result = n
    else:
        result = fib(n - 1) + fib(n - 2)
    cache[n] = result
    return result

print(fib(10))   # 55
```

```python
def send_notification(event_type: str, *args, **kwargs):
    """
    Real-world function used in backend systems for sending notifications.
    """
    timestamp = "2026-07-19 15:45:22"   # In real code, use datetime.now()
    
    # *args handling with tuple unpacking inside
    if args:
        message_parts = [str(item) for item in args]
        main_message = " | ".join(message_parts)
    else:
        main_message = "No message provided"
    
    # Build notification payload
    payload = {
        "event": event_type,
        "message": main_message,
        "timestamp": timestamp,
        "priority": kwargs.get("priority", "normal"),
        "user_id": kwargs.get("user_id"),
        "metadata": kwargs
    }
    
    # In production this would send to Slack, Email, Push, etc.
    print(f"[NOTIFICATION] [{event_type.upper()}] {main_message}")
    
    return payload


# ==================== USAGE EXAMPLES ====================

# Basic usage
send_notification("server_started")

# Using tuple unpacking at runtime (very common in prod)
event_data = ("Database backup completed", "Size: 2.3 GB", "Duration: 45s")
send_notification("backup_success", *event_data)

# Multiple arguments + extra info
send_notification(
    "payment_failed",
    "Transaction declined",
    "Card: ending with 4242",
    "Reason: insufficient funds",
    user_id=98765,
    priority="high",
    retry_count=3
)

# Another runtime unpacking example
error_info = ("Connection timeout", "Host: api.stripe.com", "Port: 443")
send_notification("api_error", *error_info, user_id=11223, priority="critical")
```

```python
def log(level, *messages, **fields):
    parts = [f"[{level}]", *messages]
    if fields:
        parts.append(" | " + " ".join(f"{k}={v}" for k, v in fields.items()))
    print(" ".join(parts))

log("INFO", "user signed in", user="ada", ip="10.0.0.1")
# [INFO] user signed in | user=ada ip=10.0.0.1
```

```python
from typing import Sequence

def calculate_discount(
    prices: Sequence[float],
    discount_percent: float,
    apply_tax: bool = False,
) -> list[float]:
    """
    Calculate final prices after applying a discount and optional tax.

    Args:
        prices: A sequence of original prices.
        discount_percent: The discount to apply, from 0.0 to 1.0.
        apply_tax: Whether to add 8 percent tax after the discount.

    Returns:
        A list of final calculated prices.

    Raises:
        ValueError: If discount_percent is not between 0 and 1.
    """
    if not 0.0 <= discount_percent <= 1.0:
        raise ValueError("Discount must be between 0.0 and 1.0")
    tax_rate = 1.08 if apply_tax else 1.0
    return [p * (1 - discount_percent) * tax_rate for p in prices]
```

```python
final_prices = calculate_discount([3000, 6000, 12000], discount_percent=0.1)
print(final_prices)
```

```python
def sum_list(values):
    if not values:
        return 0
    return values[0] + sum_list(values[1:])

print(sum_list([1, 2, 3, 4, 5]))   # 15
```

```python
x = "global x"

def outer():
    x = "enclosing x"

    def inner():
        x = "local x"
        print(x)             # local x

    inner()
    print(x)                 # enclosing x

outer()
print(x)                     # global x
```

```python
# Global config (like a default settings file)
config = {
    "db_host": "localhost",
    "db_port": 5432,
    "timeout": 30,
}

def get_database_connection(env: str):
    # Enclosing config: environment-specific overrides
    config = {
        "db_host": "prod-db.example.com" if env == "prod" else "staging-db.example.com",
        "db_port": 5432,
        "timeout": 60 if env == "prod" else 45,
    }

    def connect_for_request(request_id: str):
        nonlocal config  # refer to the enclosing 'config', not a new local
        # Local config: request-specific override
        local_config = {
            "db_host": config["db_host"],
            "db_port": config["db_port"],
            "timeout": 10,  # short timeout for this specific request
        }

        print(f"[{request_id}] Using config:", local_config)  # local config

    connect_for_request("req-123")
    print("Env config after request:", config)  # enclosing config (unchanged)

get_database_connection("prod")
print("Global config:", config)  # global config (unchanged)
```

```python
# Global log prefix (entire application)
log_prefix = "[MyApp]"

def user_handler():
    # Enclosing prefix: this specific handler
    log_prefix = "[MyApp.UserHandler]"

    def handle_request(request_id: str):
        # Local prefix: this specific request
        log_prefix = f"[MyApp.UserHandler.{request_id}]"

        print(log_prefix, "Processing request")  # local

    handle_request("req-001")
    print(log_prefix, "Handler finished")        # enclosing

user_handler()
print(log_prefix, "App shutting down")            # global
```
